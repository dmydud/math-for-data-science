# Linear Regression

```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

## Simple Linear Regression

```python
data, target = make_regression(n_samples=1000, n_features=1,
                       bias=500, noise=10, random_state=42)

plt.plot(data, target, '.')
plt.title("Data")
plt.xlabel("independent value")
plt.ylabel("dependent value");
```

```python
reg = LinearRegression().fit(data, target)
reg.score(data, target)
```

```python
print("Intercept:", reg.intercept_)
print("Coefficients:", reg.coef_)
```

```python
plt.plot(data, target, '.')
plt.title("Data")
plt.xlabel("independent value")
plt.ylabel("dependent value")
x1, x2 = plt.xlim()
y1 = reg.intercept_ + x1 * reg.coef_[0]
y2 = reg.intercept_ + x2 * reg.coef_[0]
plt.plot([x1, x2], [y1, y2]);
```

```python
residuals = target - reg.predict(data)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].plot([-3, 3], [0, 0], c="orange")
sns.scatterplot(x=data.ravel(), y=residuals, ax=ax[0])
ax[0].set_title("Residuals plot")
sns.histplot(residuals, ax=ax[1])
ax[1].set_title("Histogram of residuals")
stats.probplot(residuals, plot=ax[2])
ax[2].set_title("Residuals Normal QQ plot");
```
